## Get twitts stream and do sentiment analysis

```python
from tweepy import Stream
from tweepy import StreamListener
import json
from  textblob import TextBlob
import re
import csv
```

```python
import nltk
nltk.download('punkt')

from nltk import word_tokenize,sent_tokenize
```

```python
ubereats = 0
zomato = 0

header_name = ['UberEats', 'Zomato']
with open('sentiment.csv', 'w') as file:
    writer = csv.DictWriter(file, fieldnames=header_name)
    writer.writeheader()
    
class Listener(StreamListener):
    def on_data(self, data):
        raw_twitts = json.loads(data)
        try:
            tweets = raw_twitts['text']
            
            tweets = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweets).split())
            tweets = ' '.join(re.sub('RT',' ', tweets).split())
            
            blob = TextBlob(tweets.strip())
            
            global ubereats
            global zomato
            
            ubereats_sentiment = 0
            zomato_sentiment = 0
            
            
            for sent in blob.sentences:
                if "UberEats" in sent and "Zomato" not in sent:
                    ubereats_sentiment = ubereats_sentiment + sent.sentiment.polarity
                else:
                    zomato_sentiment = zomato_sentiment + sent.sentiment.polarity
            
            ubereats = ubereats + ubereats_sentiment
            zomato = zomato + zomato_sentiment
            
            with open('sentiment.csv', 'a') as file:
                writer = csv.DictWriter(file, fieldnames=header_name)
                info = {
                    'UberEats': ubereats,
                    'Zomato': zomato
                }
                    
                writer.writerow(info)
            
            print(tweets)
            print()
        except:
            print('Found an Error')
    def on_error(self, status):
        print(status)
            
                
```

```python
twitter_stream = Stream(auth, Listener())
twitter_stream.filter(track = ['UberEats', 'Zomato'])
```
