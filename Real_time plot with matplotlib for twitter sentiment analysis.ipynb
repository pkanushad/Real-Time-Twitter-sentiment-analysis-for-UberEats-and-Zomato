```python
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation
%matplotlib notebook
```

```python
plt.style.use('fivethirtyeight')
```

```python
frame_len = 10000

fig = plt.figure(figsize=(9,6))

def animate(i):
    data = pd.read_csv('sentiment.csv')
    y1 = data['UberEats']
    y2 = data['Zomato']

    if len(y1)<=frame_len:
        plt.cla()
        plt.plot(y1, label='UberEats')
        plt.plot(y2, label='Zomato')
    else:
        plt.cla()
        plt.plot(y1[-frame_len: ], label='UberEats')
        plt.plot(y2[-frame_len: ], label='Zomato')
    
    plt.legend(loc='upper left')
    plt.tight_layout()


ani = FuncAnimation(plt.gcf(), animate, interval=1000)
```
